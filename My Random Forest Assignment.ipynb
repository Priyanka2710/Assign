```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import  train_test_split
from sklearn.preprocessing import StandardScaler 
```

```python
from sklearn.datasets import load_boston
boston= load_boston()
X =pd.DataFrame(boston.data,columns=boston.feature_names)
y =boston.target
```

```python
boston.feature_names
```

## Handling Null Values

```python
X.isnull().sum()
```

## Splitting Dataset Into Training And Testing 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

## Training Model On Data

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor()
regressor.fit(X_train,y_train)
```

## Predicting Test Set Result

```python
y_pred=regressor.predict(X_test)
#print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
print("Score: ",regressor.score(X_test,y_test))
```

## Applying Cross Fold Validation for better result

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X_train, y = y_train, cv = 50)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

## Applying Grid Search to find the best model and the best parameters

```python
from sklearn.model_selection import GridSearchCV
param_grid = [{'n_estimators': np.arange(100,300,50),
               "criterion" : ["mse", "mae"],
               'max_depth': np.arange(10, 50, 10) ,
               'min_samples_split': np.arange(10, 50, 10),
               'min_samples_leaf': np.arange(10, 50, 10)}]

gs = GridSearchCV(estimator=regressor, param_grid=param_grid, verbose=2)

gs = gs.fit(X_train, y_train)

print(gs.best_score_)

print(gs.best_params_)

my_model = gs.best_estimator_

my_model.fit(X_train, y_train)
```

```python
y_pred = my_model.predict(X_test)

print("Accuraccuracy:", my_model.score(X_test, y_test))

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print("R2 Score:",r2_score(y_test, y_pred))
print("mean_absolute_error:",mean_absolute_error(y_test, y_pred))
print("mean_squared_error:",mean_squared_error(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = my_model, X = X_train, y = y_train, cv = 50)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
