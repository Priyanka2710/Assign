```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
%matplotlib inline
```

```python
iris=datasets.load_iris()
#iris=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
#iris_data=iris.drop(['species'],axis=1)
#iris_data.head()''''''
```

```python
#X_reduced=PCA(n_components=3).fit_transform(iris_data)
X_reduced=PCA(n_components=3).fit_transform(iris.data)
X_reduced #reduce dimension of data as we want 3 features to extract
y=iris.target #gives prediction or target  values
fig=plt.figure(1,figsize=(8,6)) 
ax=Axes3D(fig,elev=-150,azim=110)#plot 3dplot with elevation and azimuthal angle given
ax.scatter(X_reduced[:,0],X_reduced[:,1],X_reduced[:,2],c=y,cmap=plt.cm.rainbow_r)
ax.set_title("iris tranformed 3d plot")
ax.set_xlabel("x data")
ax.set_ylabel("y data")
ax.set_zlabel("z data")
plt.show()
```
